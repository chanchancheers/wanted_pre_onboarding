# 1. Tokenizer 생성하기

```python
#예시 문장
sent1 = "I go to school"
sent2 = "I LIKE pizza!"
sent_list = [sent1, sent2]
```

```python
class Tokenizer():
    def __init__(self):
        self.word_dict = {'oov': 0}
        self.fit_checker = False
    
    def preprocessing(self, sequences):
        result = []
        '''
        문제 1-1.
        '''
        total_tokens =[]

        #정규화
        import re
        for sentence in sequences:
            sentence = re.sub(r"[^a-z ]", "", sentence.lower())
            sentence = sentence.split()
            result.append(sentence)

        return result
    
    def fit(self, sequences):
        self.fit_checker = False
        '''
        문제 1-2.
        '''
        tokens = sum(self.preprocessing(sequences), [])
        tokens = set(tokens)

        #토큰이 없으면 추가, 있으면 무시
        for token in tokens:
            if token not in self.word_dict.keys():
                i = 1
                if i not in self.word_dict.values():
                    self.word_dict[token] = i
                else :
                    while i in self.word_dict.values(): 
                        i += 1
                    self.word_dict[token] = i 

        

        self.fit_checker = True
    
    def transform(self, sequences):
        result = []
        tokens = self.preprocessing(sequences)
        if self.fit_checker:
            '''
            문제 1-3.
            '''
            sequences = self.preprocessing(sequences)
            sent = []
            for sentence in sequences:
                for token in sentence:
                    if token in self.word_dict.keys():
                        token = self.word_dict[token]
                        sent.append(token)
                    else:
                        token = self.word_dict['oov']
                        sent.append(token)
                result.append(sent)
                sent = []
        
            return result
        else:
            raise Exception("Tokenizer instance is not fitted yet.")


    def fit_transform(self, sequences):
        self.fit(sequences)
        return self.transform(sequences)
    
```
